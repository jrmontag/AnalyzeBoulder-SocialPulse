## Twitter & Tumblr firehose volume

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import json

# prettier graphs, thanks to @Cmrn_DP
s = json.load( open("./bmh_matplotlibrc.json") )
matplotlib.rcParams.update(s)
```

```python
# read in twitter data
tw = pd.read_csv('./twitter.csv'
                 , sep=','
                 , names=["publisher","datetime","count"]
                 )
# data is sampled @ 20%
tw["count"] *= 5
# have a look
tw.head()
```

```python
# munge twitter data, actually only need two columns
tw = tw[["datetime", "count"]]
tw.columns = ["datetime", "twitter"]

#tw.head()
```

```python
# read in tumblr data, smarter this time via the usecols kwarg
tu = pd.read_csv('./tumblr.csv'
                 , sep=','
                 , usecols=[1,2]
                 , names=["datetime","tumblr"]
                 )
# same sampling
tu["tumblr"] *= 5 
tu.head()
```

```python
# combine the dataframes, convert the datetimes, and create an index
both = pd.merge(tw, tu, how="outer", on="datetime")
both["datetime"] = pd.to_datetime(both["datetime"], format="%Y-%m-%d %H:%M:%S")
both = both.set_index("datetime")

both_hr = both.resample(rule="60Min", how="sum")

both.head()
both_hr.head()
```

```python
%pylab inline

plt.figure()
ax = both_hr.plot(
          style='-' 
          , markersize=10
          , alpha=0.75
          #, ylim=[0, 2.6e6]
          , xlim=[pd.to_datetime("2014-01-31 00:00:00", format='%Y-%m-%d %H:%M:%S')
                  , pd.to_datetime("2014-02-02 00:00:00", format='%Y-%m-%d %H:%M:%S') 
                  ]
          , legend=True
          , label="firehose activities"
          #, subplots=True
          , figsize=(15,5)
          )

plt.legend(loc='best')
plt.ylabel('firehose activities / 1-hr')
```

## Expected event (winter storm Janus)

```python
# read in storm data
janus = pd.read_csv('./janus.csv'
                    , sep=','
                    , names=["datetime","count"]
                    #, index_col=0
                    #, parse_dates=True
                    )
#janus.dtypes
janus.tail()
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess

lowess_line = lowess(janus["count"], janus.index, frac = 0.15)
```

```python
lowess_line[10:]
```

```python
%pylab inline

ax = janus.plot(style='-' 
                , markersize=10
                , alpha=0.75
                , figsize=(15,5)
                , ylim=[-1,50] 
                , xlim=[0,5280] 
                , legend=False
                , label="'janus' activities"
                )

ax.set_ylabel('activities / min')


plt.plot(lowess_line[:,0], lowess_line[:, 1]
         , '-'
         , lw = 3
         , alpha = 0.75
         , color = 'black'
         , label = "Lowess fit"
         )
plt.legend(loc = "upper left")
```
